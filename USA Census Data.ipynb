# Data Exploration


# Which state has the most counties in it? 

```python
# Count the number of counties in each state
county_counts = df.groupby('STNAME').size()
print(county_counts)
```

```python
# Maximum County
max_counties = county_counts.max()
print(max_counties)
```

```python
# State with the highest count of counties
state_by_max_counties = county_counts[county_counts == max_counties].index[0]
print(f'State with the most counties is {state_by_max_counties} with {max_counties} counties.')
```

```python
# Assume 'df' is your DataFrame and 'STNAME' and 'COUNTY' are the columns for state names and number of counties respectively
# Group by state name and count the number of counties in each state
state_counts = df.groupby('STNAME')['COUNTY'].count()

# Sort the states by number of counties and take the top 10
top_10_states = state_counts.sort_values(ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(10,5))
plt.bar(top_10_states.index, top_10_states.values)

# Add title and labels
plt.title('Top 10 States with the Most Counties')
plt.xlabel('State')
plt.ylabel('Number of Counties')

# Display the plot
plt.show()
```

# Considering only the top three most populated counties within each state, identify the three states with the highest population. The population data should be based on the ‘CENSUS2010POP’ field. List the states in descending order of population.

```python
# Convert the population column to integers
df['CENSUS2010POP'] = df['CENSUS2010POP'].astype(int)
df['CENSUS2010POP'].describe()
```

```python
# Group the data by state and sum the population for each state
populations = df.groupby('STNAME')['CENSUS2010POP'].sum()
populations
```

```python
# Sort the data by population in descending order
populations = populations.sort_values(ascending=False)
populations
```

```python
# Get the top 3 most populous states
top_3_states = populations.sort_values(ascending=False).head(3)
top_3_states
print(f'The Top 3 most populous states are: {top_3_states.index[0]} ({top_3_states[0]:,}), {top_3_states.index[1]} ({top_3_states[1]:,}), and {top_3_states.index[2]} ({top_3_states[2]:,}).')
```

```python
# Group by state name and sum the population in each state, then take the top 3
top_3_populations = df.groupby('STNAME')['CENSUS2010POP'].sum().nlargest(3)

# Create a horizontal bar chart
plt.figure(figsize=(10,5))
plt.barh(top_3_populations.index, top_3_populations.values)

# Add title and labels
plt.title('Top 3 States by Population')
plt.xlabel('Population')
plt.ylabel('State')

# Display the plot
plt.show()
```

# Which County has the most Cities in it?

```python
# Count the number of counties in each city
county_counts_city = df.groupby('CTYNAME').size()
print(county_counts_city)
```

```python
#Find the maximum county number in a city
max_counties_city = county_counts_city.max()
print(max_counties_city)
```

```python
# Find the city with the maximum number of counties
city_with_max_counties = county_counts_city[county_counts_city == max_counties_city].index[0]
print(f'The County with the most Cities is {city_with_max_counties} with {max_counties_city} counties.')
```

```python
# Group by county name and count the number of cities in each county, then take the top 5
top_5_counties = df.groupby('CTYNAME').size().nlargest(5)

# Create a pie chart
plt.figure(figsize=(10,5))
plt.pie(top_5_counties.values, labels=top_5_counties.index, autopct=lambda p: '{:.1f}% ({:.0f})'.format(p, p * sum(top_5_counties.values) / 100))

# Add title
plt.title('Top 5 Counties with the Most Cities')

# Display the plot
plt.show()
```

# Which region has the most divisions in it?

```python
# Count the number of counties in each state
division_counts = df.groupby('REGION').size()
print(division_counts)
```

```python
# Find the maximum division
max_division = division_counts.max()
print(max_division)
```

```python
# Find the Region with the maximum number of divisions
region_with_max_division = division_counts[division_counts == max_division].index[0]
print(f'The region with the most divisions is Region {region_with_max_division} with {max_division} divisions.')
```

```python
# Group by region and count the number of occurrences in each region
division_counts = df.groupby('REGION').size()

# Convert the index to string type
division_counts.index = division_counts.index.astype(str)

# Create a line chart
plt.figure(figsize=(10,5))
plt.plot(division_counts.index, division_counts.values)

# Add title and labels
plt.title('Regions by Divisions')
plt.xlabel('Region')
plt.ylabel('Count')

# Display the plot
plt.show()
```
